```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

cv2.line(image, start, end, color, thickness)

선 그리기

```python
# 선 그리기
img = np.full((512,512, 3), 255, np.uint8)
img = cv2.line(img, (0,0),(255,255), (255,0,0), 10)
#img = cv2.line(img, (0,0),(255,255), (255,0,0), 3)
plt.imshow(img)
plt.show()
```

cv2.rectangle(image, start,end,color,thickness)

```python
# 직사각형 그리기
img = np.full((512,512,3), 255, np.uint8)
img = cv2.rectangle(img, (40,40), (300,300), (255,200,0), 10)
plt.imshow(img)
plt.show()

print()
img = cv2.rectangle(img, (40,40), (300,300), (255,200,0), -1) # -1의 경우 채우기
plt.imshow(img)
plt.show()
```

cv2.circle(image, center, radian, color, thickness)

```python
# 원 그리기
img = np.full((512,512,3), 255, np.uint8)
img = cv2.circle(img, (200, 200), 100, (255,0,100), -1)
plt.imshow(img)
plt.show()

img = np.full((512,512,3), 255, np.uint8)
img = cv2.circle(img, (200, 200), 100, (255,0,100), -1)
img = cv2.circle(img, (450, 450), 100, (100,100,100), -1)
plt.imshow(img)
plt.show()
```

cv2.polylines(image, points, is_close, color, thickness)

```python
# 다각형 그리기
img = np.full((512, 512, 3), 255, np.uint8)
points = np.array([[5, 5], [128, 258], [483, 444], [400, 150]])
img =cv2.polylines(img, [points], True, (0,0,255), 4)

plt.imshow(img)
plt.show()

img = np.full((512, 512, 3), 255, np.uint8)
points = np.array([[5, 5], [128, 258],[300, 450], [483, 444], [400, 150] ])
img =cv2.polylines(img, [points], True, (0,0,255), 5)

plt.imshow(img)
plt.show()
```

cv2.putText(image, text, position, font_type, font_scale, color)

```python
# 글자 그리기
img = np.full((512,512,3), 255, np.uint8)
img = cv2.putText(img, "KingNamJi", (100,200), cv2.FONT_ITALIC,2 ,(255,0,0))

plt.imshow(img)
plt.show()
```
