```python
import cv2
import matplotlib.pyplot as plt
```

외곽을 찾아주는 함수

cv2.findContours(image, mode, method)

mode (contour)들을 찾는 방법
- RETR_EXTERNAL : 바깥쪽 line 만
- RETR_LIST: 모든 Line을 찾지만 hierarchy 구성 x
- RETR_TREE: 모든 Line을 찾으며, 모든 hierarchy 구성

method (contour)들을 찾는 근사치 방법
- CHAIN_APPROX_NONE: 모든 contour 포인트 저장
- CHAIN_APPROX_SIMPLE: Contour Line을 그릴 수 있는 포인트만 저장

cv2.drawContours(image, contours, contour_index, color, thickness) Contour 들을 그리는 함수

```python
img = cv2.imread('/content/drive/MyDrive/opencv/cat.jpeg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
#plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR))
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
ret, thresh = cv2.threshold(img_gray, 127, 255, 0)
plt.imshow(thresh,cmap = 'gray')
plt.show()
```

```python
# 바깥쪽 라인만 찾도록
contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
img_contours = cv2.drawContours(img, contours, -1, (0,255,0), 4)

plt.imshow(cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB))
```

```python
# 모든 라인을 찾도록
contours = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[0]
img_contours = cv2.drawContours(img, contours, -1, (0,255,0), 4)

plt.imshow(cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB))
```

```python
# 모든 라인을 찾도록2
contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[0]
img_contours = cv2.drawContours(img, contours, -1, (0,255,0), 4)

plt.imshow(cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB))
```
