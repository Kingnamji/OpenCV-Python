```python
img = cv2.imread('/content/drive/MyDrive/opencv/cat.jpeg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# resize
expand = cv2.resize(img, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_CUBIC) 
plt.imshow(cv2.cvtColor(expand, cv2.COLOR_BGR2RGB))
plt.show() # x축 y축을 통해 이미지가 커진 것을 확인
```

```python
# 이미지 위치 변경
img = cv2.imread('/content/drive/MyDrive/opencv/cat.jpeg')

height, width = img.shape[:2] # 행과 열의 크기만 저장

M = np.float32([[1, 0, 50],[0, 1, 10]]) # 변환을 위한 행렬을 만들어 준다.
dst = cv2.warpAffine(img, M, (width, height))

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 이미지 위치 변경
M = np.float32([[1, 0, 100],[0, 1, 50]]) # 변환을 위한 행렬을 만들어 준다.
dst = cv2.warpAffine(img, M, (width, height))

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

getRotationMatrix2D(center, angle, scale)를 사용하면 회전 이미지 회전을 위한 변환 행렬을 반환할 수 있다.

```python
# 이미지의 회전
img = cv2.imread('/content/drive/MyDrive/opencv/cat.jpeg')

height, width = img.shape[:2] # 행과 열의 크기만 저장
M = cv2.getRotationMatrix2D((width/2, height / 2), -90, 0.8)
dst = cv2.warpAffine(img, M, (width, height))
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
