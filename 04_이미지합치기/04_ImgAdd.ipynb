```python
# 라이브러리
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread('/content/drive/MyDrive/opencv/cat.jpeg')
img2 = cv2.imread('/content/drive/MyDrive/opencv/gangajipool.png')

plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

cv2.add(): Saturation 연산을 수행
- Saturation: 0보다 작으면 0, 255보다 크면 255로 표현

np.add(): Modulo 연산(256)을 수행

```python
# cv2.add() 는 크기가 같은 이미지에 대한 연산만 가능
height, width = img1.shape[:2]

#resize를 통해 크기를 맞춰 줌
dst = cv2.resize(img2, dsize=(width, height), interpolation=cv2.INTER_AREA) 
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
print(img1.shape)
print(dst.shape)
result1 = cv2.add(img1, dst)
plt.imshow(cv2.cvtColor(result1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# np.add()
result2 = img1 + dst
plt.imshow(cv2.cvtColor(result2, cv2.COLOR_BGR2RGB))
plt.show()
```
