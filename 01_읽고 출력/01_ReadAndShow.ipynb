코랩에서는 OpenCV의 imshow

```python
import cv2
import matplotlib.pyplot as plt
```

- imread 는 이미지를 읽어 numpy 객체로 만든다.

- openCV의 경우 numpy 객체를 반환할 때
(행, 열, 색상 BGR 순)

- imshow를 통해 윈도우 창에 띄우는게 불가능하기 때문에, matplotlib를 사용해 출력

```python
img_1 = cv2.imread('/content/drive/MyDrive/opencv/cat.jpeg', cv2.IMREAD_COLOR) # 사진을 [B, G, R] 순의 numpy 객체를 반환
plt.imshow(cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)) # matplotlib : [R, G, B]
plt.show()
```

```python
img_gray = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY) # 흑백으로 변환
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB))
plt.show()
```
