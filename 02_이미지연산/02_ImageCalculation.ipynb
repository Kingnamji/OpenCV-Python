```python
# 라이브러리 불러오기
import cv2
import matplotlib.pyplot as plt
import time
```

```python
img = cv2.imread('/content/drive/MyDrive/opencv/cat.jpeg')
print(img.shape) # 이미지 크기 확인 + 흑백, 컬러 유무
print(img.size) # 총 픽셀 수
```

```python
px = img[100,100]
print(px) # 100*100 위치 픽셀 B G R 순을 출력
print(px[2]) # R 값
print(px[1]) # G 값
print(px[0]) # B 값 
```

```python
# 특정 범위 픽셀 변경, 시간 측정
start = time.time()
for i in range(0,200):
    for j in range(0,200):
        img[i,j] = [255, 255, 255]
print(f'걸린 시간: {time.time()-start}s')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
start = time.time()
img[0:200, 0:200] = [0, 0, 0]
print(f'걸린 시간: {time.time()-start}s')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# ROI (관심 영역) 추출
roi = img[400:750, 250:600]
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# ROI 단위로 이미지 복사
img[0:350, 0:350] = roi
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 픽셀별 색상 값 조정
img = cv2.imread("/content/drive/MyDrive/opencv/cat.jpeg")
img[:,:, 2] = 0 # 전체 픽셀 R 색상 값 0으로

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img = cv2.imread("/content/drive/MyDrive/opencv/cat.jpeg")
img = img[:,:, :]//2  # 전체 픽셀 색상 값 어둡게
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
